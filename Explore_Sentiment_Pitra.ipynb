```python
import pandas as pd
import openai
from google.colab import files

# Load the CSV file into a DataFrame
file_path = 'normalized_comments.csv'
df = pd.read_csv(file_path)

# Print the column names to confirm
print(df.columns)

# Set your OpenAI API key
openai.api_key = "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" # Replace with your OpenAI API key

# Define a function to analyze context relevance using OpenAI API
def analyze_relevance(text, model='gpt-3.5-turbo'):
    prompt = f"Analyze the following text and determine if it is related to context a coffee shop or coffee or drink or food or barista or location expansion or taste or support. Classify it as 'Yes' (if any correlation), 'No' (if no one correlation):\n\n{text}"

    response = openai.ChatCompletion.create(
        model=model,
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]
    )

    relevance_response = response['choices'][0]['message']['content'].strip()

    # Extract relevance classification from the response
    if "Yes" in relevance_response:
        relevance = "Yes"
    elif "No" in relevance_response:
        relevance = "No"
    else:
        relevance = "Unknown"

    return relevance

# Apply the relevance analysis function to each normalized comment
df['Correlation'] = df['Normalized_Comment'].apply(lambda x: analyze_relevance(x))

# Save the resulting DataFrame to a new CSV file
df.to_csv('context_relevance_comments.csv', index=False)

# Provide a download link for the new CSV file
files.download('context_relevance_comments.csv')
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import math
from scipy import stats

# Load the data from Excel file
file_path = 'Final_Result_Pitra_v1.xlsx'  # Adjust the file path as needed
data = pd.read_excel(file_path)

# Extract the confidence scores
confidence_scores = data['Confidence_Score']

# Calculate Mean
mean_confidence = confidence_scores.mean()
print(f"Mean Confidence Score: {mean_confidence}")

# Calculate Median
median_confidence = confidence_scores.median()
print(f"Median Confidence Score: {median_confidence}")

# Calculate Mode
mode_confidence = confidence_scores.mode()[0]
print(f"Mode Confidence Score: {mode_confidence}")

# Calculate Standard Deviation
std_dev_confidence = confidence_scores.std()
print(f"Standard Deviation of Confidence Score: {std_dev_confidence}")

# Calculate Margin of Error
n = len(confidence_scores)
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error = z_score * (std_dev_confidence / math.sqrt(n))
print(f"Margin of Error: {margin_of_error}")

# Print summary
print("\nSummary:")
print(f"Mean Confidence Score: {mean_confidence}")
print(f"Median Confidence Score: {median_confidence}")
print(f"Mode Confidence Score: {mode_confidence}")
print(f"Standard Deviation of Confidence Score: {std_dev_confidence}")
print(f"Margin of Error: {margin_of_error}")

# Additional Insight: Distribution of Sentiments
sentiment_distribution = data['Sentiment'].value_counts()
print("\nSentiment Distribution:")
print(sentiment_distribution)

# Additional Insight: Distribution of Correlation
correlation_distribution = data['Correlation'].value_counts()
print("\nCorrelation Distribution:")
print(correlation_distribution)
```
